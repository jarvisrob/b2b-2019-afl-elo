```r
# Functions
normal_pdf_pts <- function(mu, sigma, n_pts) {
    n_sigmas <- 4
    x <- seq(from = mu - 4 * sigma, to = mu + 4 * sigma, length.out = n_pts)
    pdf <- dnorm(x, mean = mu, sd = sigma)
    pdf_pts <- tibble(x, pdf)
}
```

## He chose an arbitrary scale
- "Average" player was rated 1500
- Standard deviation of players' performance was 200 on the same scale

```r
player_avg <- 1500
player_sd <- 200

player1_rating <- player_avg + player_sd  # One std dev above average
player2_rating <- player_avg  # Average
player1_pdf <- normal_pdf_pts(player1_rating, player_sd, 1000)
player2_pdf <- normal_pdf_pts(player2_rating, player_sd, 1000)
ggplot() + 
    geom_line(data = player1_pdf, mapping = aes(x, pdf), color = "blue") + 
    geom_line(data = player2_pdf, mapping = aes(x, pdf), color = "red")
```

## This implies probabilities
- If during any one game Player 1's performance is better than Player 2, they win
- Based on the plot above, you'd expect that more often than otherwise
- But due to random variation, Player 2 might outperform Player 1 on some occassions

```r
delta_ratings <- player1_rating - player2_rating
combined_player_sd <- sqrt(2 * player_sd ^ 2)
combined_player_pdf <- normal_pdf_pts(delta_ratings, combined_player_sd, 1000)
combined_player_pdf %>% ggplot(aes(x, pdf)) + geom_line(colour = "black")
cat(str_c("P(Player 1 defeats Player 2) = P(X > 0) = ", pnorm(0, delta_ratings, combined_player_sd, lower.tail = FALSE)))
```

## But, the rating scale is arbitrary!
What matters is the **difference** in rating between two players **relative** to the chosen standard deviation.

By how many standard deviations is one player rated above another?

By rating players appropriately, with a **difference that is in the correct proportion to the standard deviation**, you can reflect the **probability** that one player will defeat the other.

```r
for (delta in seq(from = -400, to = 400, by = 50)) {
    cat(str_c("Diff = ", delta, ": P(Player 1 d. Player 2) = ", pnorm(0, delta, combined_player_sd, lower.tail = FALSE), "\n"))
}

# This would be better as a plot
```

## Which leads us to the important question ...
## How do you rate players?

### The answer is *iteratively*
- Adjust a player's rating after **each game**, either up or down
- Adjustment is based on their performance relative to expectations
- **Expectation** is equal to the probabilty that a player will win

$$R_A'=R_A+K(S_A-E_A)$$
where
- $R_A'=$ new rating for Player A
- $R_A=$ old rating for Player A
- $K=$ adjustment coefficient
- $S_A=$ score, being 1 if Player A wins, 0 if they lose, 0.5 if they draw
- $E_A=$ expectation, the probability of Player A winning

Note that when Player A vs Player B:

$$S_B=1-S_A$$

and

$$E_B=1-E_A$$

which means that

$$R_B'=R_B+K(S_B-E_B)=R_B-K(S_A-E_A)$$

Points are transferred from the player that underperformed expectations to the player that overperformed expectations, by an amount proportional to the difference between the actual result and the expected result:

$$K(S_A-E_A)$$

### The Elo rating algorithm is continually tuning itself based on the last result
*Run through an example, use $K=20$*

## The next question is then ... what is the correct value of $K$?
- A big value of $K$ means ratings change rapidly, and more closely reflect recent **form**
- A small value of $K$ means ratings change slowly, and more closely align with **talent**

### Tuning $K$ is **key** to making Elo's rating system work
