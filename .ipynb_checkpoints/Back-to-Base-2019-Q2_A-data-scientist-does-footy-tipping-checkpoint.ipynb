# Rating chess players

```r
player_avg <- 1500
player_sd <- 200

player1_rating <- player_avg + player_sd  # One std dev above average
player2_rating <- player_avg  # Average

player1_pdf <- normal_pdf_curve(player1_rating, player_sd, 1000)
player2_pdf <- normal_pdf_curve(player2_rating, player_sd, 1000)

ggplot() + 
    geom_line(data = player1_pdf, mapping = aes(x, pdf), color = "blue") + 
    geom_line(data = player2_pdf, mapping = aes(x, pdf), color = "purple")
```

## This means we can make probabilistic forecasts

```r
delta_ratings <- player1_rating - player2_rating
combined_player_sd <- sqrt(2 * player_sd ^ 2)

combined_player_pdf <- normal_pdf_curve(delta_ratings, combined_player_sd, 1000)

combined_player_pdf %>% 
    ggplot(aes(x, pdf)) + 
        geom_line(colour = "red") + 
        geom_vline(xintercept = 0, color = "black")

cat(str_c("P(Player 1 wins) = P(X > 0) = ", pnorm(0, delta_ratings, combined_player_sd, lower.tail = FALSE)))
```

```r
player_cdf <-
    tibble(
        delta = seq(-600, 600, by = 5),
        p_win = pnorm(0, delta, combined_player_sd, lower.tail = FALSE)
    )

player_cdf %>% ggplot(aes(delta, p_win)) + geom_line(color = "blue")
```

## If the rating scale and the uncertainty (standard deviation) in performance are arbitrary, then what matters is *the ratings you give to players*
## So how do you rate players?

### The answer is *iteratively*: Keep updating based on new information

- Adjust a player's rating after **each game**, either up or down
- Adjustment is based on their performance relative to expectations
- **Expectation** is equal to the probabilty that a player will win

$$R_A'=R_A+K(S_A-E_A)$$
where
- $R_A'=$ new rating for Player A
- $R_A=$ old rating for Player A
- $K=$ adjustment coefficient
- $S_A=$ score, being 1 if Player A wins, 0 if they lose, 0.5 if they draw
- $E_A=$ expectation, the probability of Player A winning

Note that when Player A vs Player B:

$$S_B=1-S_A$$

and

$$E_B=1-E_A$$

which means that

$$R_B'=R_B+K(S_B-E_B)=R_B-K(S_A-E_A)$$

**In effect, points are transferred from the player that underperformed expectations to the player that overperformed expectations. The amount transferred is proportional to the difference between the actual result and the expected result:**

$$K(S_A-E_A)$$

### The Elo rating algorithm is continually tuning itself based on the last result
*Run through an example, use $K=20$*

## The big question is then ... What is the correct value of $K$?
- A big value of $K$ means ratings change rapidly, and more closely reflect recent **form**
- A small value of $K$ means ratings change slowly, and more closely align with **talent**

## Tuning $K$ is **key** to making Elo's rating system work

# From Grandmasters to Grand Finals

```r
hawks_rating <- +18
cats_rating <- -12
team_sd <- 38 / sqrt(2)

hawks_pdf <- normal_pdf_curve(hawks_rating, team_sd, 1000)
cats_pdf <- normal_pdf_curve(cats_rating, team_sd, 1000)

ggplot() +
    geom_line(hawks_pdf, mapping = aes(x, pdf), color = "brown") +
    geom_line(cats_pdf, mapping = aes(x, pdf), color = "blue")
```

```r
margin_expected <- hawks_rating - cats_rating
margin_sd <- 38

margin_pdf <- normal_pdf_curve(margin_expected, margin_sd, 1000)

margin_pdf %>% 
    ggplot(aes(x, pdf)) + 
        geom_line(color = "red") + 
        geom_vline(xintercept = 0, color = "black")

cat(str_c("P(Hawks win) = P(Margin > 0) = ", pnorm(0, margin_expected, margin_sd, lower.tail = FALSE)))
```

```r
margin_actual <- 50

margin_actual_pdf <- normal_pdf_curve(margin_actual, margin_sd, 1000)

margin_pdf %>% 
    ggplot() + 
        geom_line(margin_pdf, mapping = aes(x, pdf), color = "red") + 
        geom_line(margin_actual_pdf, mapping = aes(x, pdf), color = "blue") +
        geom_vline(xintercept = 0, color = "black")

EA <- pnorm(0, margin_expected, margin_sd, lower.tail = FALSE)
SA <- pnorm(0, margin_actual, margin_sd, lower.tail = FALSE)

cat(str_c("EA = P(Hawks win) = P(X > 0) = ", EA, "\n"))
cat(str_c("SA = P(X > 0 with margin = ", margin_actual, ") = ", SA, "\n"))
cat(str_c("SA - EA = ", SA - EA, "\n"))
```

## Our scale and the uncertainty (standard deviation) are no longer arbitrary--they are linked to the predicted margin and the uncertainty in the margin

But it's circular:
- Need to know $\sigma$ to predict the margin
- Need to know the predicted margin to calculate $\sigma$

## That's the basics of a Elo rating system for AFL, but we can do more

### Regression to the mean between seasons
- Biggest unknowns tend to be at the start of a season
- All teams are **regressed to the mean** before the start of a new season
- Regression amount is controlled by the distance from the mean rating, i.e. $0$, and parameter $\alpha$

$$R_A' = R_A - \alpha(R_A-0)=(1-\alpha)R_A$$


### Ground ratings for each team
- For each team on each ground
- Iterative adjustment of team performance **on that ground**

Update after each game by amount:

$$G(S_A-E_A)$$

### It's hard to play on the road because Australia is *big*

$$T=C_1d^{C_2}$$

where

- $T=$ travel penalty
- $d=$ distance traveled
- $C_1=$ travel penalty coefficient, with $C_1\geq0$
- $C_2=$ travel penalty exponent, with $0{\leq}C_2{\leq}1$

```r
c1 <- 1.1485622
c2 <- 0.3170362

travel_penalty <- travel_curve(c1, c2, 1000)

travel_penalty %>% ggplot(aes(d, T)) + geom_line(color = "blue")
```

## It becomes an optimisation problem

Need to tune the values of **seven** parameters:
- $\sigma$ to match the observed uncertainty in predicting margins
- $K=$ the coefficient for updating ratings
- $\alpha=$ the regression fraction between seasons
- $G=$ the coefficient for ground rating updates
- $C_1$ and $C_2$, the travel penalty parameters
- $R_{expn,0}=$ the initial rating for expansion teams

with the goal of making the probabilities and margins accurate, and the margin as precise as possible

# Evaluating performance
## Tune on historical data 2007-2017, check against 2018

```r
season_2018 <-
    season_2018_elo_data %>%
        select(
            team.home, 
            team.away, 
            delta.rating.home, 
            margin.act.home,
            margin.error.home,
            result.exp.home, 
            outcome.home,
            result.exp.away, 
            outcome.away, 
            brier.game
        ) %>%
        rename(
            home = team.home,
            away = team.away,
            margin_expected = delta.rating.home,
            margin_actual = margin.act.home,
            margin_error = margin.error.home,
            p_home_win = result.exp.home,
            outcome_home = outcome.home,
            p_away_win = result.exp.away,
            outcome_away = outcome.away,
            brier = brier.game,
        )

season_2018
```

```r
# Calibration curve
season_2018_home <- season_2018 %>% select(p_home_win, outcome_home) %>% rename(p = p_home_win, outcome = outcome_home)
season_2018_away <- season_2018 %>% select(p_away_win, outcome_away) %>% rename(p = p_away_win, outcome = outcome_away)
season_2018_pred_act <- union(season_2018_home, season_2018_away)

n_preds <- season_2018_pred_act %>% nrow()

season_2018_pred_act %>%
    group_by(cut(p, breaks = seq(0, 1, 0.05))) %>%
    summarize(n(), sum(outcome)) %>%
    rename(bin = `cut(p, breaks = seq(0, 1, 0.05))`, n_games = `n()`, n_wins = `sum(outcome)`) %>%
    mutate(bin_centre = seq(0.075, 0.925, 0.05), win_fraction = n_wins / n_games) %>%
    ggplot(aes(bin_centre, win_fraction)) +
        geom_point() +
        geom_abline(slope = 1, intercept = 0) +
        scale_x_continuous(limits = c(0, 1))

season_2018_pred_act %>% filter(p < 0.2)
```

```r
# Measures of performance
cat(str_c("Margin mean absolute error = ", mean(abs(season_2018$margin_error)), " points\n"))
cat(str_c("Margin error std dev = ", sd(season_2018$margin_error), " points\n"))
cat(str_c("Brier score = ", mean(season_2018$brier), "\n"))

# Tipping score
n_correct_tips <-
    season_2018 %>%
        filter((p_home_win > 0.5 & outcome_home == 1) | (p_away_win > 0.5 & outcome_away == 1)) %>%
        count() %>%
        pull(n)
n_games <- season_2018 %>% count()
cat(str_c("Tipping score = ", n_correct_tips, " out of ", n_games, " (including finals) = ", n_correct_tips/n_games*100, " per cent\n"))
```

### 20,000 Monte Carlo simulations later

```r
team_of_interest <- "hawthorn"
```

```r
n_itns <- 20000

ladder_posn_prob <- 
  ladder_mc_sims %>%
    filter(team == team_of_interest) %>%
    group_by(ladder.posn) %>%
    count() %>%
    mutate(p = n/n_itns)

ladder_posn_prob %>%
    ggplot(aes(ladder.posn, p)) + 
      geom_col() +
      ggtitle(
        paste0(
          "Probability of end ladder position for ", 
          team_of_interest
        )
      ) +
      xlab("Ladder position") +
      ylab("Probability")

tibble(
    event = c(
        "Top of ladder (Minor premiers)",
        "Top 4 (2nd chance in finals)",
        "Top 8 (Play in finals)",
        "Bottom of ladder (Wooden spoon)"
    ),
    p = c(
        ladder_posn_prob %>% filter(ladder.posn == 1) %>% pull(p),
        ladder_posn_prob %>% filter(ladder.posn <= 4) %>% pull(p) %>% sum(),
        ladder_posn_prob %>% filter(ladder.posn <= 8) %>% pull(p) %>% sum(),
        ladder_posn_prob %>% filter(ladder.posn == 16) %>% pull(p)
    )
)
```
